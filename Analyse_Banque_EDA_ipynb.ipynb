```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Columns you want to analyze
columns_to_analyze = [
    'Estimated Income',
    'Superannuation Savings',
    'Credit Card Balance',
    'Bank Loans',
    'Bank Deposits',
    'Checking Accounts',
    'Saving Accounts',
    'Foreign Currency Account',
    'Business Lending'
]

# Function to detect outliers using IQR
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Dictionary to store results
outlier_counts = {}

# Loop through all columns in the list
for col in columns_to_analyze:
    outliers = detect_outliers_iqr(df, col)
    outlier_counts[col] = {
        'Outlier Count': len(outliers),
        'Outlier %': (len(outliers) / len(df)) * 100
    }

    # Boxplot visualization
    plt.figure(figsize=(6, 3))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot - {col}")
    plt.show()

# Create summary table
outlier_summary = pd.DataFrame(outlier_counts).T
display(outlier_summary.sort_values(by='Outlier Count', ascending=False))
```

### Key Observations:

- **Saving Accounts**, **Bank Deposits**, and **Checking Accounts** have the highest number and percentage of outliers, ranging from about 4.6% to 5.2% of the data. This indicates that a significant portion of values in these categories deviate substantially from the norm, which may be due to extreme account balances or data entry errors.

- **Business Lending**, **Credit Card Balance**, **Bank Loans**, and **Foreign Currency Account** have moderate outlier percentages, between approximately 2.6% and 3.1%. These could represent unusually large loans or credit balances.

- **Estimated Income** and **Superannuation Savings** have relatively few outliers, less than 1% of the data, suggesting that these variables are more stable or have fewer extreme values.
